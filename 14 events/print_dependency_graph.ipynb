```python
import pandas as pd
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_trf")

# Path to your csv
csv_file = "../7 Dataset/prem_vs_conc_csv_files/all/A2008_Commission of the European Communities v Salzgitter AG.csv"  

# Read the CSV
df = pd.read_csv(csv_file)
```

```python

# Loop over each text row and show the dependency parse
for idx, row in df.iterrows():
    text = row['text']
    label = row['label']
    print(f"\n---\nText #{idx+1} ({label}):\n{text}\n")
    doc = nlp(text)

    # Display dependency parse in notebook (uncomment inner line if in notebook)
    displacy.render(doc, style="dep", jupyter=True, options={'compact': True, 'distance': 90})
    # # For script: create an SVG file per row for later review
    # svg = displacy.render(doc, style="dep", options={'compact': True, 'distance': 90})
    # out_filename = f"dep_vis_{idx+1}_{label}.svg"
    # with open(out_filename, "w", encoding="utf-8") as f:
    #     f.write(svg)
    # print(f"Dependency visualization for this text saved as {out_filename}")

    # Optionally: Print tokens with their dependency labels
    print("Token dependencies:")
    for token in doc:
        print(
            f"{token.text:<15} {token.dep_:<10} "
            f"Head: {token.head.text:<15} "
            f"POS: {token.pos_:<8}"
        )
```
